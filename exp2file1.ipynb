```python
import numpy as np
```

```python
np.random.seed (1)
```

```python
x= np.random.randint(0,50,1000)
```

```python
y=x+np.random.normal(0,10,1000)
y=x+np.random.normal(0,10,1000)
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline
plt.scatter(x,y)
plt.show()
```

```python
y=100-x+np.random.normal(0,10,1000)
```
